```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as sts
import numpy as np
```

```python
## 2015 Data Cleaning
# Import 2015 World Happiness Data
data_2015 = pd.read_csv('./Resources/2015_data.csv')

# Add column indicating the year
data_2015['Year'] = '2015'

# Drop non-predictive variables
data_2015_drop = data_2015.drop(columns='Standard Error')

# Rearrange the data fields
column_order = ['Year','Country','Region','Happiness Rank','Happiness Score','Economy (GDP per Capita)','Family',
                'Health (Life Expectancy)','Freedom','Trust (Government Corruption)','Generosity','Dystopia Residual']
data_2015_clean = data_2015_drop.reindex(columns = column_order)

# Trim the data set to only show top 20 countries
#data_2015_clean = data_2015_order.drop(range(20,len(data_2015)))

data_2015_clean
```

```python
## 2016 Data Cleaning
# Import 2016 World Happiness Data
data_2016 = pd.read_csv('./Resources/2016_data.csv')

# Add column indicating the year
data_2016['Year'] = '2016'

# Drop non-predictive variables
data_2016_drop = data_2016.drop(columns=['Lower Confidence Interval','Upper Confidence Interval'])

# Rearrange the data fields
column_order = ['Year','Country','Region','Happiness Rank','Happiness Score','Economy (GDP per Capita)','Family',
                'Health (Life Expectancy)','Freedom','Trust (Government Corruption)','Generosity','Dystopia Residual']
data_2016_clean = data_2016_drop.reindex(columns = column_order)

data_2016_clean
```

```python
## 2017 Data Cleaning
# Import 2017 World Happiness Data
data_2017 = pd.read_csv('./Resources/2017_data.csv')

# Add column indicating the year
data_2017['Year'] = '2017'

# Rename data fields to match prior years
data_2017_rename_1 = data_2017.rename(columns={'Happiness.Rank': 'Happiness Rank'})
data_2017_rename_2 = data_2017_rename_1.rename(columns={'Happiness.Score': 'Happiness Score'})
data_2017_rename_3 = data_2017_rename_2.rename(columns={'Economy..GDP.per.Capita.': 'Economy (GDP per Capita)'})
data_2017_rename_4 = data_2017_rename_3.rename(columns={'Health..Life.Expectancy.': 'Health (Life Expectancy)'})
data_2017_rename_5 = data_2017_rename_4.rename(columns={'Trust..Government.Corruption.': 'Trust (Government Corruption)'})
data_2017_rename_6 = data_2017_rename_5.rename(columns={'Dystopia.Residual': 'Dystopia Residual'})                          

# Drop non-predictive variables
data_2017_drop = data_2017_rename_6.drop(columns=['Whisker.high','Whisker.low'])

# Rearrange the data fields
column_order = ['Year','Country','Happiness Rank','Happiness Score','Economy (GDP per Capita)','Family',
                'Health (Life Expectancy)','Freedom','Trust (Government Corruption)','Generosity','Dystopia Residual']
data_2017_clean = data_2017_drop.reindex(columns = column_order)

data_2017_clean
```

```python
## 2018 Data Cleaning
# Import 2018 World Happiness Data
data_2018 = pd.read_csv('./Resources/2018_data.csv')

# Add column indicating the year
data_2018['Year'] = '2018'

# Rename data fields to match prior years
data_2018_rename_1 = data_2018.rename(columns={'Overall rank': 'Happiness Rank'})
data_2018_rename_2 = data_2018_rename_1.rename(columns={'Country or region': 'Country'})
data_2018_rename_3 = data_2018_rename_2.rename(columns={'Score': 'Happiness Score'})
data_2018_rename_4 = data_2018_rename_3.rename(columns={'GDP per capita': 'Economy (GDP per Capita)'})
data_2018_rename_5 = data_2018_rename_4.rename(columns={'Social support': 'Family'})
data_2018_rename_6 = data_2018_rename_5.rename(columns={'Healthy life expectancy': 'Health (Life Expectancy)'})
data_2018_rename_7 = data_2018_rename_6.rename(columns={'Freedom to make life choices': 'Freedom'})
data_2018_rename_8 = data_2018_rename_7.rename(columns={'Perceptions of corruption': 'Trust (Government Corruption)'})
                            
# Rearrange the data fields
column_order = ['Year','Country','Happiness Rank','Happiness Score','Economy (GDP per Capita)','Family',
                'Health (Life Expectancy)','Freedom','Trust (Government Corruption)','Generosity','Dystopia Residual']
data_2018_clean = data_2018_rename_8.reindex(columns = column_order)

data_2018_clean
```

```python
## 2019 Data Cleaning
# Import 2019 World Happiness Data
data_2019 = pd.read_csv('./Resources/2019_data.csv')

# Add column indicating the year
data_2019['Year'] = '2019'

# Rename data fields to match prior years
data_2019_rename_1 = data_2019.rename(columns={'Overall rank': 'Happiness Rank'})
data_2019_rename_2 = data_2019_rename_1.rename(columns={'Country or region': 'Country'})
data_2019_rename_3 = data_2019_rename_2.rename(columns={'Score': 'Happiness Score'})
data_2019_rename_4 = data_2019_rename_3.rename(columns={'GDP per capita': 'Economy (GDP per Capita)'})
data_2019_rename_5 = data_2019_rename_4.rename(columns={'Social support': 'Family'})
data_2019_rename_6 = data_2019_rename_5.rename(columns={'Healthy life expectancy': 'Health (Life Expectancy)'})
data_2019_rename_7 = data_2019_rename_6.rename(columns={'Freedom to make life choices': 'Freedom'})
data_2019_rename_8 = data_2019_rename_7.rename(columns={'Perceptions of corruption': 'Trust (Government Corruption)'})
                            
# Rearrange the data fields
column_order = ['Year','Country','Happiness Rank','Happiness Score','Economy (GDP per Capita)','Family',
                'Health (Life Expectancy)','Freedom','Trust (Government Corruption)','Generosity','Dystopia Residual']
data_2019_clean = data_2019_rename_8.reindex(columns = column_order)

data_2019_clean
```

```python
data_list = [data_2015_clean, data_2016_clean, data_2017_clean, data_2018_clean, data_2019_clean]

multi_year_data = pd.concat(data_list)

multi_year_data
```

```python
total_country_rank = multi_year_data.groupby('Country')['Happiness Rank'].sum()

avg_country_rank = total_country_rank/5

data_rank = pd.merge(left=multi_year_data, right=avg_country_rank, on='Country')
data_rank
```

```python
# Trim the data set to only show top 20 countries
# data_rank_top = data_rank.drop(range(100,len(data_rank)))

# Rename average happiness rank
data_rank_top_rename_1 = data_rank.rename(columns={'Happiness Rank_x': 'Happiness Rank'})
data_rank_top_rename_2 = data_rank_top_rename_1.rename(columns={'Happiness Rank_y': 'Avg Happiness Rank'})

data_rank_top_rename_2
```

```python
data_rank_top_rename_2.to_csv('Resources/clean_data/data_country_rankings_all_years.csv', 
                              index=False)
```
